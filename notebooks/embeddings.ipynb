```python
s = "this is a test"
```

```python
nlp = spacy.load("en_vectors_web_lg")
```

```python
#%%timeit
tokens = nlp(s)
```

```python
for tok in tokens:
    print("token: {}\nvector: {}\nnorm: {}\nis out of vocabulary: {}\n".format(tok.text, tok.vector, tok.vector_norm, tok.is_oov))
```

```python
nlp_bert = spacy.load("en_trf_bertbaseuncased_lg")
```

```python
%%timeit
tokens = nlp_bert(s)
```

```python
for tok in tokens:
    print("token: {}\nvector: {}\nnorm: {}\nis out of vocabulary: {}\n".format(tok.text, tok.vector, tok.vector_norm, tok.is_oov))
```

```python
nlp_roberta = spacy.load("en_trf_robertabase_lg")
```

```python
%%timeit
tokens = nlp_roberta(s)
```

```python
for tok in tokens:
    print("token: {}\nvector: {}\nnorm: {}\nis out of vocabulary: {}\n".format(tok.text, tok.vector, tok.vector_norm, tok.is_oov))
```
